# Python Modules and Packages
A Python **package** is a collection of software written in Python. Many Python packages are *open-source*, meaning they are developed and released for free online. 

A package can have one of more **modules** in it. A module is a single Python file, often with functions that you can import and use in your own code as long as you have the package installed.

A package that you will encounter often when doing scientific coding with Python is **matplotlib**. This is a plotting package that offers many modules for producing publication-grade figures. One of the *modules* in matplotlib is **pyplot**, and this is the main plotting module we will use in this lesson.

## Importing modules

Any installed packages can be made available for use in a given script with an `import` statement. By convention, we write all imports at the top of our Python programs. Modules within the package are invoked with a dot `.` after the package name. Functions are invoked with a dot `.` after the name of the module they are found in.

```python
import matplotlib # Imports the package matplotlib

from matplotlib import pyplot # Imports the pyplot module in the matplotlib package
x = (0, 1, 2, 3, 4)
y = (0, 1, 2, 4, 8)

matplotlib.pyplot.plot(x, y) # The plot function in the pyplot module
```

Using the dots makes it explicit where your functions are located. Future you or your collaborators can go back to the code above and know immediately that the `plot` function can be found in a file called `pyplot.py` in the `matplotlib` package. 

The problem is that these long funtion names can be unwieldy and make your code more confusing than necessary. Python offers two ways to deal with this. One way is more encouraged than the other. Try to think about which style is better in the following two examples.

```python
from matplotlib.pyplot import * # Imports ALL functions and variables from matplotlib
plot(x, y)
```

```python
import matplotlib.pyplot as plt # Define the short form plt for matplotlib.pyplot
plt.plot(x, y)
```

Both types of imports seem to do the same thing, and `import *` made it more convenient to simply type `plot()`. But should we always use `import *`? 

Imagine a collaborator wrote a module called `coolplotlib` that nicely plots your data. What happens if you want to plot your data with `matplotlib`'s `plot` and `coolplotting`'s `plot`?

```python
import matplotlib.pyplot as mplt
import coolplotlib as cplt

subplot(1, 2, 1) # Set up a side by side plot
mplt.plot(x, y)

subplot(1, 2, 2) # The second subplot
cplt.plot(x, y)
```

```python
from matplotlib.pyplot import *
from coolplotlib import *
plt.subplot(1, 2, 1) 
plot(x, y)

plt.subplot(1, 2, 2) 
plot(x, y)
```

Whoops. When we `import *`, Python cannot distinguish the `matplotlib.plot()` function from the `coolplotlib.plot()` function. If functions or variables have duplicate names when you `import *`, Python will simply overwrite them. To avoid this problem, it is best to use the `module.function()` convention. This convention also makes it explicit where your functions are coming from. This is especially useful for future you or collaborators trying to understand what functions you are using and which modules they can be found in.
